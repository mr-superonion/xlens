```python
def angular_area_in_arcminutes_from_pixel(pixel, pixel_scale):
    # Calculate the area in pixel space (in square pixels)
    pixel_area = np.pi * pixel**2

    # Convert pixel area to angular area in square arcseconds
    angular_area_arcsec2 = pixel_area * (pixel_scale**2)

    # Convert from square arcseconds to square arcminutes
    angular_area_arcmin2 = angular_area_arcsec2 / 3600

    return angular_area_arcmin2


def angular_area_in_arcminutes(radius_arcmin):
    # Convert the radius from arcminutes to radians
    radius_rad = radius_arcmin * np.pi / (180 * 60)

    # Calculate the angular area in steradians using the formula
    area_sr = 2 * np.pi * (1 - np.cos(radius_rad))

    # Convert the angular area from steradians to square arcminutes
    area_arcmin2 = area_sr * (10800 / np.pi) ** 2  # (10800 = 180 * 60)

    return area_arcmin2
```

```python
fig, axes = plt.subplots(6, 2, figsize=(24, 26))
angular_bin_mid = (
    (
        np.mean(summary_table["angular_bin_left"], axis=0)
        + np.mean(summary_table["angular_bin_right"], axis=0)
    )
    / 2
    / 60
)


mean_eX = np.mean(summary_table["eX"], axis=0)
mean_rX = np.mean(summary_table["rX"], axis=0)
sigma_eX = np.std(summary_table["eX"].to_numpy(), axis=0) / np.sqrt(
    len(summary_table)
)
sigma_rX = np.std(summary_table["rX"].to_numpy(), axis=0) / np.sqrt(
    len(summary_table)
)
sigma_gX = np.sqrt(
    (sigma_eX / mean_rX) ** 2 + (mean_eX * sigma_rX / mean_rX**2) ** 2
)

true_gX = np.mean(summary_table["r_weighted_gX"], axis=0) / np.mean(
    summary_table["rX"], axis=0
)

axes[0, 0].errorbar(
    angular_bin_mid, mean_eX / mean_rX, label="measured gX", yerr=sigma_gX
)
axes[0, 0].plot(
    angular_bin_mid,
    true_gX,
    label="r weighted true gX",
)
axes[0, 0].set_ylim(-0.01, 0.01)
axes[0, 0].set_ylabel("gX")
axes[0, 0].set_xlabel("arcmin")
axes[0, 0].legend()

axes[1, 0].plot(angular_bin_mid, mean_eX / mean_rX - true_gX, label="true gX")
axes[1, 0].set_ylabel(r"gX_meas - gX_true")

mean_eT = np.mean(summary_table["eT"], axis=0)
mean_rT = np.mean(summary_table["rT"], axis=0)
sigma_eT = np.std(summary_table["eT"].to_numpy(), axis=0) / np.sqrt(
    len(summary_table)
)
sigma_rT = np.std(summary_table["rT"].to_numpy(), axis=0) / np.sqrt(
    len(summary_table)
)
sigma_gT = np.sqrt(
    (sigma_eT / mean_rT) ** 2 + (mean_eT * sigma_rT / mean_rT**2) ** 2
)
true_gT = np.mean(summary_table["r_weighted_gT"], axis=0) / np.mean(
    summary_table["rT"], axis=0
)

axes[0, 1].errorbar(
    angular_bin_mid, mean_eT / mean_rT, label="measured gT", yerr=sigma_gT
)

axes[0, 1].set_ylim(0, 0.05)
# axes[0,1].plot(angular_bin_mid, summary_table["gT_true"]/ summary_table["rT"],label="true gT")
axes[0, 1].plot(
    angular_bin_mid,
    true_gT,
    label="r weighted true gT",
)
axes[0, 1].legend()
axes[0, 1].set_xlabel("arcmin")
axes[0, 1].set_ylabel("gT")

axes[1, 1].plot(
    angular_bin_mid, mean_eT / mean_rT / true_gT - 1, label="true gT"
)
axes[1, 1].set_ylabel(r"gT_meas/gT_true - 1")
axes[1, 1].set_ylim(-0.04, 0.04)

# axes[1,0].errorbar()

axes[3, 0].errorbar(
    angular_bin_mid,
    np.mean(summary_table["kappa_true"].to_numpy(), axis=0)
    / np.mean(summary_table["ngal_in_bin"].to_numpy(), axis=0),
    yerr=np.std(summary_table["kappa_true"].to_numpy(), axis=0)
    / np.sqrt(len(summary_table))
    / np.mean(summary_table["ngal_in_bin"].to_numpy(), axis=0),
)
axes[3, 0].set_ylabel(rf"$<\kappa$>")
axes[3, 0].set_xlabel("arcmin")

axes[2, 0].errorbar(
    angular_bin_mid,
    np.mean(summary_table["rX"].to_numpy(), axis=0)
    / np.mean(summary_table["ngal_in_bin"].to_numpy(), axis=0),
    yerr=np.std(summary_table["rX"].to_numpy(), axis=0)
    / np.sqrt(len(summary_table))
    / np.mean(summary_table["ngal_in_bin"].to_numpy(), axis=0),
)
axes[2, 0].set_xlabel("arcmin")
axes[2, 0].set_ylabel(rf"<rX>")
# axes[1, 0].set_ylim(0.16, 0.185)

axes[2, 1].errorbar(
    angular_bin_mid,
    np.mean(summary_table["rT"].to_numpy(), axis=0)
    / np.mean(summary_table["ngal_in_bin"].to_numpy(), axis=0),
    yerr=np.std(summary_table["rT"].to_numpy(), axis=0)
    / np.sqrt(len(summary_table))
    / np.mean(summary_table["ngal_in_bin"].to_numpy(), axis=0),
)
axes[2, 1].set_xlabel("arcmin")
axes[2, 1].set_ylabel(rf"<rT>")
# axes[1, 1].set_ylim(0.16, 0.185)


axes[3, 1].errorbar(
    angular_bin_mid,
    np.mean(summary_table["ngal_in_bin"].to_numpy(), axis=0) / area,
    yerr=np.std(summary_table["ngal_in_bin"].to_numpy(), axis=0)
    / np.sqrt(len(summary_table))
    / area,
    label=f"n_halo={len(summary_table)}",
)
axes[3, 1].set_xlabel("arcmin")
axes[3, 1].set_ylabel(rf"Detection Number density [arcmin$^{-2}$]")
axes[3, 1].legend()

axes[4, 0].errorbar(
    angular_bin_mid,
    np.mean(summary_table["lensed_shift"], axis=0),
    yerr=np.std(summary_table["lensed_shift"].to_numpy(), axis=0)
    / np.sqrt(len(summary_table)),
    label="shift",
)
axes[4, 0].errorbar(
    angular_bin_mid,
    np.mean(summary_table["radial_lensed_shift"], axis=0),
    yerr=np.std(summary_table["radial_lensed_shift"].to_numpy(), axis=0)
    / np.sqrt(len(summary_table)),
    label="radial_shift",
)
axes[4, 0].set_ylabel("Lens shift [arcsec]")
axes[4, 0].legend()

axes[4, 1].plot(
    angular_bin_mid,
    np.mean(summary_table["median_match_dist"], axis=0),
    label="match failure rate",
)
axes[4, 1].set_ylabel("Median match distance pixel")

axes[5, 0].plot(
    angular_bin_mid,
    np.mean(summary_table["match_failure_rate"], axis=0),
    label="match failure rate",
)
axes[5, 0].set_ylabel("Match failure rate (> 2 pixel)")

for ax in axes.flatten():
    ax.set_xlim(0, 8)
    ax.set_xlabel("Prelensed angular separation [arcmin]")

plt.subplots_adjust(hspace=0.2, wspace=0.2)
```
